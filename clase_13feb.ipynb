## Ciclo while

Este lo usamos cuando no sabemos exactamente cuánto durará un proceso. Hay que tener cuidado de verificar que alguna condición eventualmente se cumpla para que el programa no entre en un bucle infinito.

```python
cota = 100
a, b = 1, 1
while b <= cota:
    a, b = b, a + b
    print ("b = ", b)
```

## Ciclo if

```python
x, y = 7, 3
if x < y:
    print(x, ' es menor que ', y)
elif x==y:
    print(x, ' son iguales')
else:
    print(x, 'es mayor que ', y)
```

```python
#Pongmoes nuestro coseno
def kos(x, n_max = 8): 
    y = 0
    x = x%(2*pi) #Tomamos el módulo de las cosas
    for n in range (n_max):
        y = y + ((-1)**(n))*((x**(2*n))/(factorial(2*n))) #Estamos usando la serie de McLaurin hasta el vigésimo término.
        # Podemos escribir y += lo que vamos a sumar
    return y
```

```python
kos(pi/4)
```

```python
%%timeit
cos(pi/4)
```

```python
kos(100*pi)
```
