## Primera función del curso 

De cálculo sabemos que:

$$
\cos (x) = \sum_{n=0}^{\infty}{-1^{n}\frac{x^{2n}}{(2n)!}} = 1 - \frac{x^{2}}{2} + \frac{x^4}{24} + \dots
$$

Python trabaja las listas desde 0. Entonces, se crea en el intervalo [0,n).

A continuación, usamos la serie de **McLaurin** para desarrollar la expresión del coseno hasta el vigésimo término:

```python
#Definimos la función coseno
def kos(x, n_max = 5): 
    y = 0
    for n in range (n_max):
        y = y + ((-1)**(n))*((x**(2*n))/(mt.factorial(2*n))) #Estamos usando la serie de McLaurin hasta el vigésimo término.
        # Podemos escribir y += lo que vamos a sumar
    return y
```

Vemos cuál es el recíproco de la raíz cuadrada de 2:

```python
1/mt.sqrt(2) #Esta es la raíz cuadrada de 2
```

Vemos cuál es el coseno de pi/4

```python
kos(mt.pi/4) #Usamos la función con el coseno de pi/4
```

```python
mi_cos = kos(x, 5)
```

```python
plt.plot(x,cos_tradicional, '--')
plt.plot(x,mi_cos, '-')
plt.ylim([-1.1,1.1])
plt.xlim([x.min(), x.max()])

plt.show()
```

Veamos la comparación de las gráficas

```python
plt.figure(figsize = (10,6))
plt.plot(x, cos_tradicional, '--', lw = 5) #Muestra la gráfica original

for i in range(1,10):
    plt.plot(x,kos(x,i), label = '$N_{max}= %1.0f$' %i) #Grafica la correspondiente
    
plt.ylim([-1.1,1.1])
plt.xlim([x.min(), x.max()])
plt.legend(loc=3)
plt.title('Gráficas del coseno y de sus aproximaciones con serie de McLaurin.')
plt.xlabel('Argumento (x)')
plt.ylabel('x')
plt.show()
```
